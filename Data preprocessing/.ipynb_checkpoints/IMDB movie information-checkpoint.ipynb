## 8. Handling the null values(Drop / Fill).

```python
#Droping null values:
df.drop(columns=['homepage', 'belongs_to_collection'], inplace=True, errors='ignore')
```

```python
df['tagline'] = df['tagline'].fillna('No tagline')
df['Keywords'] = df['Keywords'].fillna('keyword not found')
df['production_companies'] = df['production_companies'].fillna('companies unknown')
df['production_countries'] = df['production_countries'].fillna('countries unknown')
df['spoken_languages'] = df['spoken_languages'].fillna('unknown')
df['crew'] = df['Keywords'].fillna('not found')
df['cast'] = df['cast'].fillna('no cast')
df['overview'] = df['overview'].fillna('overview not found')
df['runtime'] = df['runtime'].fillna('no runtime')
df['poster_path'] = df['poster_path'].fillna('not found')
df.isnull().sum()
```

## 12. Extract the genres (note: some movies may have multiple genres, split them).

```python
from collections import Counter
import ast
```

```python
def extract_genre(genres_str):
    try:
        genres = ast.literal_eval(genres_str)
        return [genre['name'] for genre in genres]
    except(SyntaxError, ValueError):
        return ['unknown']
df['genres_list'] = df['genres'].apply(extract_genre)
all_genres = [genre for sublist in df['genres_list'] for genre in sublist]
genre_counts = Counter(all_genres)
genre_counts_df = pd.DataFrame(genre_counts.items(), columns = ['Genre','Count'])
for genre, count in genre_counts.items():
    print(f'{genre}:{count}')
```

## 13. Number of movies in each genre.

```python
#Number of movies in each genre:
genre_counts = df['genres'].value_counts()
print(genre_counts)
```

## 14. Bar chart of the movie counts.

```python
#Bar chart of the movie counts per genre
plt.figure(figsize=(10,6))
plt.bar(genre_counts_df['Genre'],genre_counts_df['Count'], color = 'pink', edgecolor='black' ) 
plt.xticks(rotation=45, ha='right')  
plt.title('Movie counts per genre.')
plt.xlabel('Genre Name')
plt.ylabel('Number of movies in each genre')
plt.tight_layout()
plt.show()
```

**Observation:**   
The bar chart titled “Movie counts per genre” this shows how many movies fall into each genre.
 1) The x-axis lists different genres.
 2) The y-axis shows the number of movies in each genre.
 3) Taller bars indicate genres with more movies, shorter bars indicate fewer movies.
