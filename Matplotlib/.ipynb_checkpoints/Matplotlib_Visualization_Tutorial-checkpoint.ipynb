## Importing Required Libraries

```python
#Importing Required Libraries (Matplotlib and NumPy)
import matplotlib.pyplot as plt
import numpy as np
```

## Line Plot

```python
#Simple Line Plot
x = [1,2,3,4]
y = [2,4,6,8]
plt.plot(x,y)
plt.show()
```

**Observation:**   
This plot shows a straight line where y increases proportionally with x.

## Line Plot with Axis Labels and Title

```python
#Line Plot with Axis Labels and Title
x = [1,2,3,4]
y = [2,4,6,8]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Table')
plt.show()
```

**Observation:**  
This line plot shows `y` increasing proportionally to `x` with slope 2.

## Customized Line Plot

```python
#Customized Line Plot (Color, Linestyle, Markers, Figure Size)
plt.figure(figsize=(10,6))
plt.plot(x,y, color = 'darkred', linestyle = '--', marker = 'o')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Table')
plt.show()
```

**Observation:**   
The line is customized with color, linestyle, and markers for better visualization.

## Bar Chart

```python
#Bar Chart for Fruit Sales
fruits  = ['Apple','Orange','Banana']
sales = [10,20,30]
plt.figure(figsize = (5,5))
plt.bar(fruits,sales, color = 'teal')
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.show()
```

**Observation:**   
Bar chart displays the sales quantity for different fruits.

## Bar Chart

```python
#Bar Chart for Students' Marks
students = ['Anu','Rekha','Rathi','Reena']
marks = [99,88,66,77]
plt.figure(figsize = (3,3))
plt.bar(students,marks, color = 'teal')
plt.xlabel('Students Name')
plt.ylabel('Marks')
plt.title('Students List')
plt.show()
```

**Observation:**   
Bar chart displays the marks scored by each student.

## Basic Pie Chart

```python
#Basic Pie Chart
name = ['A','B','C','D']
sizes = [10,40,20,30]
plt.pie(sizes, labels=name, autopct='%1.1f%%')
plt.show()
```

**Observation:**  
Pie chart shows the percentage distribution among four categories A, B, C, D.

## Pie Chart with Custom Colors

```python
#Pie Chart with Custom Colors
name = ['A','B','C','D']
sizes = [10,40,20,30]
col = ['#0000FF', '#006400', '#FFC0CB','#00FFFF']
plt.pie(sizes, labels = name, colors = col, autopct = '%1.1f%%')
plt.show()
```

**Observation:**   
Pie chart shows the same distribution with custom colors applied.

## Pie Chart with Seaborn Color Palette and Equal Aspect Ratio

```python
#Pie Chart with Seaborn Color Palette and Equal Aspect Ratio
category = ['Category A','Category B','Category C','Category D']
num = [10,40,20,30]
import seaborn as sns
color = sns.color_palette('magma',len(num))
plt.figure(figsize = (3,3))
plt.pie(num, labels = name, colors = color, autopct = '%1.1f%%', startangle= 90)
plt.axis('equal')
plt.show()
```

**Observation:**   
Pie chart uses a seaborn color palette and equal aspect ratio to look more balanced

## Scatter Plot

```python
#Scatter Plot Example
x =  [2,4,6,8,10]
y = [22,33,66,55,44]
plt.scatter(x,y, color= 'red')
plt.show()
```

**Observation:**   
Scatter plot shows the relationship between x and y values as discrete points.

## Horizontal Bar Chart

```python
#Horizontal Bar Chart for Students' Marks
stu = ['Anu','Reena','Rekha']
mark = [88,99,77]
plt.barh(stu,mark, color = 'red')
plt.show()
```

**Observation:**  
 1) This horizontal bar chart displays the marks of each student.   
 2) Reena scored the highest (99), Anu scored 88, and Rekha scored 77.


## Histogram of Numerical Data

```python
#Histogram of Numerical Data
num = [20,4,5,22,33,44,37,42,29]
plt.hist(num, bins =5, color='skyblue', edgecolor = 'red')
plt.show()
```

**Observation:**  
1) This histogram groups the `num` values into 5 bins.   
2) It visually shows how many values fall within each range, making it easy to see the distribution of the data.


## Box Plot

```python
#Box Plot of Data
a = [1,3,5,7,9,10,2,8,6,4,2,3,7,6,5,4,3]
plt.boxplot(a)
plt.show(a)
```

**Observation:**  
This box plot summarizes the distribution of values in `a`, showing the median, quartiles, and any potential outliers. It quickly highlights the central tendency and spread of the data.
