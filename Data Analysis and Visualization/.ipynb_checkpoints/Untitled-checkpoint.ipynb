# III) Data Wrangling

## 1) Checking missing/null values

```python
#Count of Missing Entries:
x = df.isnull().sum()
print(x)
```

## 2) Handling missing values

### i) Filling Missing Year:

```python
#Filling Missing Year
df['Year'] = df['Year'].fillna(0).astype(int)
df.isnull().sum()
```

### ii) Filling Missing Publisher Values:

```python
#Filling Missing Publisher
df['Publisher'] = df['Publisher'].fillna('Unknown')
df.isnull().sum()
```

# IV) Exploratory Data Analysis

## 1) Average Global Sales per Game

```python
#Average Global Sales per Game
avg_sales = df['Global_Sales'].mean()
print("Average Global Sales:", avg_sales)
```

## 2) Year with the Highest Number of Game Releases

```python
# Year with the Highest Number of Game Releases
top_year = df['Year'].value_counts().idxmax()
count_top_year = df['Year'].value_counts().max()
print("Year with most games:", top_year, "(",count_top_year,"games )")
```

## 3) Number of Unique Genres in the Dataset

```python
#Bar Chart of Genre Counts
unique_genres = df['Genre'].nunique()
print("Number of unique genres:", unique_genres)
```

## 4) Bar Chart Genres with the Most Games Released

```python
#Genres with the Most Games Released
genre_counts = df['Genre'].value_counts()
plt.figure(figsize=(10,5))
genre_counts.plot(kind='bar', color='orange')
plt.title('Distribution of Games by Genre')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

**Observation:**   
The bar chart shows that Action games have been released the most, followed by Sports and Miscellaneous genres. Genres such as Puzzle and Strategy have the fewest releases. This indicates that the gaming industry focuses heavily on Action and Sports titles compared to niche genres.

## 5) Bar Chart of Platforms with the Most Games

```python
#Platforms with the Most Games
plt.figure(figsize=(10,5))
sns.countplot(y=df['Platform'], order=df['Platform'].value_counts().head(10).index, color = 'palegreen', edgecolor = 'black')
plt.title("Top 10 Platforms by Number of Games")
plt.show()
```

**Observation:**   
Platforms such as DS, PS2, and PS3 dominate in the number of games released. This indicates that developers tend to focus on the most popular gaming consoles.

## 6) Publishers by Number of Games Released

```python
# Top 10 Publishers by Number of Games Released
publisher_counts = df['Publisher'].value_counts().head(10)
print(publisher_counts)
```

## 7) Publisher with the Highest Total Global Sales

```python
#Top 10 Publishers by Total Global Sales
publisher_sales = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)
print(publisher_sales)
```

## 8) Lollipop Chart of Publishers by Total Global Sales

```python
# Top 10 publishers by global sales
plt.figure(figsize=(12,6))
markerline, stemlines, baseline = plt.stem(publisher_sales.index, publisher_sales.values, basefmt=" ")
plt.setp(markerline, marker='o', markersize=10, color='orange')
plt.title('Top 10 Publishers by Total Global Sales', fontsize=16)
plt.xlabel('Publisher', fontsize=12)
plt.ylabel('Total Global Sales (in millions)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**Observation:**   
The lollipop chart displays the top 10 publishers by total global sales. Each “lollipop” represents a publisher, with the length of the stem showing the magnitude of sales. [Highest publisher] clearly leads the market, while the remaining publishers show a descending trend. This visualization makes it easy to compare publishers’ performance at a glance, highlighting both market leaders and the relative differences among the top players.

## 9) Bar Chart of Publishers by Total Global Sales

```python
#Bar Chart of Top 10 Publishers by Total Global Sales
publisher_sales = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
publisher_sales.plot(kind='bar', color='skyblue')
plt.title('Top 10 Publishers by Total Global Sales', fontsize=16)
plt.xlabel('Publisher', fontsize=12)
plt.ylabel('Total Global Sales (in millions)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**Observation:**   
The bar chart shows the top 10 publishers by total global sales. It is clear that a few publishers dominate the market, with [highest publisher name] leading significantly in global sales. Other publishers show a gradual decrease in total sales, indicating a competitive but skewed distribution where the top few publishers generate the majority of global revenue. The chart highlights the market leaders and their relative performance compared to others.

## 10) Share of Global Sales by Game Genre

```python
#Share of Global Sales by Game Genre
genre_sales = df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,8))
plt.pie(genre_sales, labels=genre_sales.index, autopct='%1.1f%%', startangle=140)
plt.title("Share of Global Sales by Genre")
plt.show()
```

**Observation:**   
The pie chart displays the share of global sales by game genre. It shows that certain genres contribute a much larger portion of total sales, indicating their popularity and market dominance. For example, genres like [insert top genre from data] occupy the largest segment, while others have smaller shares, highlighting the diversity in consumer preferences. This visualization helps identify which genres are driving the majority of global revenue in the gaming industry.
